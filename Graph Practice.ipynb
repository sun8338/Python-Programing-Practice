#그래프 핵심 코드 구현

* 그래프 정의



```python
class Graph():
    def __init__ (self, size):
        self.SIZE=size
        self.graph=[[0 for _ in range(size)] for _ in range(size)]
```

```python
G1, G3=None, None #전역변수선언
```

```python
G1=Graph(4)
G1.graph[0][1]=1;G1.graph[0][2]=1;G1.graph[0][3]=1
G1.graph[1][0]=1;G1.graph[1][2]=1
G1.graph[2][0]=1;G1.graph[2][1]=1;G1.graph[2][3]=1
G1.graph[3][0]=1;G1.graph[3][2]=1
```

```python
print('## G1 무방향 그래프 ##')
for raw in range(4): 
    for col in range(4):
        print(G1.graph[raw][col], end='')
    print()
```

* 방향 그래프

```python
G3=Graph(4)
G3.graph[0][1]=1; G3.graph[0][2]=1
G3.graph[3][0]=1; G3.graph[3][2]=1
```

```python
print('## G3 Directed Graph ##')
for raw in range(4): 
    for col in range(4):
        print(G3.graph[raw][col], end='')
    print()
```

#깊이 우선 탐색

- 그래프 정의

```python
class Graph():
    def __init__ (self, size):
        self.SIZE=size
        self.graph=[[0 for _ in range(size)] for _ in range(size)]
```

- 그래프 선언

```python
##전역 변수 선언 부분##
G1=None
stack=[]
visitedAry=[] #방문한 정점

##메인 코드 부분##
G1=Graph(4)
G1.graph[0][2]=1; G1.graph[0][3]=1
G1.graph[1][2]=1
G1.graph[2][0]=1; G1.graph[2][1]=1; G1.graph[2][3]=1
G1.graph[3][0]=1; G1.graph[3][2]=1
```

```python
print('## G1 Undirected Graph ##')
for raw in range(4): 
    for col in range(4):
        print(G1.graph[raw][col], end='')
    print()
```

- 깊이 우선 탐색

```python
current=0 
stack.append(current)
visitedAry.append(current)

while (len(stack)!=0): 
    next=None
    for vertex in range(4):
        if G1.graph[current][vertex]==1:
            if vertex in visitedAry:
                pass
            else:
                next=vertex
                break
    if next !=None:
        current=next
        stack.append(current)
        visitedAry.append(current)
    else:
        stack.pop()
        if len(stack) != 0 :
            current = stack[-1]
```

- 방문한 노드 살펴 보기

```python
print('방문순서-->', end=' ')
for i in visitedAry:
    print(chr(ord('A')+i), end=' ')
```
