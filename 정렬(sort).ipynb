선택 정렬 함수 구현

```python
def selectionSort(array, size):
    for step in range(size): #바깥쪽 for문, step이 0, 1, 2이런 식으로 바뀌게 됨. 
        min_idx=step #iteration을 돌때마다 최솟값으로 가정하는 인덱스
        for i in range(step +1, size): #안쪽 for문
            if array[i]<array[min_idx]: #매 iteration 마다 최솟값을 찾음. 
                min_idx=i
        (array[step], array[min_idx])=(array[min_idx], array[step]) #최솟값을 해당 자리에 교체
```

```python
data=[20, 12, 10, 15, 2]
size=len(data)
selectionSort(data, size)
print('선택 정렬 적용후 데이터')
print(data)
```

삽입정렬

```python
def insertionSort(array):

    for step in range(1, len(array)):
        key=array[step]
        j=step-1
        while j >= 0 and key < array[j]:
            array[j+1]=array[j]
            j=j-1
        array[j+1]=key
```

```python
data=[50, 9, 5, 1, 4, 3, 0]
insertionSort(data)
print('삽입 정렬 적용 후 데이터')
print(data)
```
